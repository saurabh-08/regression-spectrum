Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Loading the data
df = pd.read_csv("HW2_linear_data.csv")
X = df.iloc[:, 0].values
Y = df.iloc[:, 1].values

# Initializing values for the parameters
m, c = 0, 0
learning_rate = 0.0001
epochs = 1000

# Mean Squared Error (MSE) loss fn.
def mse(Y, Y_predicted):
    return np.mean((np.power((Y - Y_predicted), 2)))

# Performing Gradient Descent
for epoch in range(epochs):
    Y_predicted = m * X + c
    m_gradient = (-2 / len(X)) * sum(X * (Y - Y_predicted))
    c_gradient = (-2 / len(X)) * sum(Y - Y_predicted)
    m = m - learning_rate * m_gradient
    c = c - learning_rate * c_gradient

    # Loss calculation for monitoring and observation
    loss = mse(Y, Y_predicted)
    if (epoch + 1) % 50 == 0:
        print(f'Epoch: {epoch+1}/{epochs}, Loss: {loss}')

# Ploting linear regression line with matplotlib
plt.scatter(X, Y, label='Data Entries')
plt.plot(X, m*X + c, color='red', label='Linear Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression')
plt.grid(True)
plt.show()
```

Non-linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Loading data from CSV file
df = pd.read_csv('HW2_nonlinear_data.csv')
X = df['X'].values
Y = df['Y'].values

# Mean squared error (MSE) loss function for calculating the loss
def mse(Y_true, Y_pred):
    return np.mean((Y_true - Y_pred)**2)

# Defining the given cubic function model
def cubic_model(X, a, b, c, d):
    return a * X**3 + b * X**2 + c * X + d

# Initializing the model parameter with random values
a, b, c, d = np.random.rand(4)
learning_rate = 1e-6 # assigning value as per the assignment guidelines
epochs = 10000 # assigning value as per the assignment guidelines

# Gradient Descent
for epoch_i in range(epochs):
    # Calculating predictions using the current parameters
    predicted_val = cubic_model(X, a, b, c, d)

    # Calculating the gradient of the loss for each parameter separately
    gradient_a = -2 * np.mean((Y - predicted_val) * X**3)
    gradient_b = -2 * np.mean((Y - predicted_val) * X**2)
    gradient_c = -2 * np.mean((Y - predicted_val) * X)
    gradient_d = -2 * np.mean(Y - predicted_val)

    # Updating the parameters using the gradient and learning rate
    a = a - learning_rate * gradient_a
    b = b - learning_rate * gradient_b
    c = c - learning_rate * gradient_c
    d = d - learning_rate * gradient_d

    # Calculating and printing the MSE loss
    loss = mse(Y, predicted_val)
    if epoch_i % 500 == 0:
        print(f"Epoch {epoch_i}: Loss = {loss:.6f}")

# Printing the final parameter values
print("Final parameters:")
print(f"a = {a:.7f}, b = {b:.7f}, c = {c:.7f}, d = {d:.7f}")

# Creating a range of X, Y values for plotting
X_range = np.linspace(min(X), max(X), 100)
Y_range = cubic_model(X_range, a, b, c, d)

# Ploting the data points & the regression curve
plt.scatter(X, Y, label='Data points', marker='.', color='blue')
plt.scatter(X_range, Y_range, label='Cubic Regression', color='red',marker='.')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non Linear Regression')
plt.legend()
plt.show()
```
